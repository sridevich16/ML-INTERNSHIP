```python
p75=df["Electric Range"].quantile(0.75)
p25=df["Electric Range"].quantile(0.25)
print(p25,p75)
```

```python
iqr=p75-p25
print(iqr)
```

```python
ul=p75+1.5*iqr
lp=p25-1.5*iqr
print(ul,lp)
```

```python
df[df["Electric Range"]>ul].count()
```

```python
new_df=df[df["Electric Range"]<ul]
sns.boxplot(x="Electric Range",data=new_df)
```

```python
sns.histplot(new_df['Electric Range'], bins=10, kde=True)
```

```python
df[df["Electric Range"]<lp].count()
```

```python
df_cap = df.copy()
df_cap['Electric Range'] = np.where(df_cap['Electric Range'] > ul,ul,df_cap['Electric Range'])
```

```python
sns.boxplot(x="Electric Range",data=df_cap)
```
