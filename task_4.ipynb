```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Used Car Dataset.csv")
```

```python
df["mileage(kmpl)"].fillna(18.9,inplace=True)
df["engine(cc)"].fillna(df["engine(cc)"].value_counts().idxmax(),inplace=True)
df["max_power(bhp)"].fillna(df["max_power(bhp)"].value_counts().idxmax(),inplace=True)
df["torque(Nm)"].fillna(df["torque(Nm)"].value_counts().idxmax(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df.hist(figsize=(14,10))
```

```python
sns.heatmap(df[df.columns].corr(),annot=True)
```

```python
x=df[["mileage(kmpl)","kms_driven"]]
y=df["price(in lakhs)"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
LR = LinearRegression()
LR.fit(x_train, y_train)
```

```python
y_pred = LR.predict(x_test)
y_pred
```

```python
from sklearn.metrics import mean_absolute_error

y_pred = LR.predict(x_test)
MAE = mean_absolute_error(y_test, y_pred)
print(" Mean Absolute Error is : ", MAE)
```

```python
from sklearn.metrics import mean_squared_error

y_pred = LR.predict(x_test)
MSE = mean_squared_error(y_test, y_pred)
print(" Mean Squared Error is : ", MSE)
```

```python
import math
RMSE = math.sqrt(MSE)
print(" Root Mean Squared Error is : ", RMSE)
```

```python
from sklearn.metrics import r2_score

y_pred = LR.predict(x_test)
R2 = r2_score(y_test, y_pred)
print(" R2 Score is : ", R2)
```

```python

plt.scatter(y_test, y_pred, alpha=0.5,label='Actual vs Predicted', color='blue')


plt.title('Actual vs Predicted Prices')

```

```python
LR.score(x_test,y_test)
```

```python
LR.score(x_train,y_train)
```

```python
x=df[["seats","kms_driven","mileage(kmpl)","engine(cc)","max_power(bhp)","torque(Nm)"]]
y=df["price(in lakhs)"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
from sklearn.linear_model import LinearRegression
LR1 = LinearRegression()
LR1.fit(x_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error

y_pred = LR1.predict(x_test)
MAE = mean_absolute_error(y_test, y_pred)
print(" Mean Absolute Error is : ", MAE)
```

```python
LR1.score(x_test,y_test)
```

```python
LR1.score(x_train,y_train)
```

```python
from sklearn.linear_model import Lasso, Ridge
lasso_model = Lasso(alpha=0.1)  
lasso_model.fit(x_train, y_train)
```

```python
y_pred_lasso = lasso_model.predict(x_test)
```

```python
mse_lasso = mean_squared_error(y_test, y_pred_lasso)
r2_lasso = r2_score(y_test, y_pred_lasso)
```

```python
ridge_model = Ridge(alpha=50)
ridge_model.fit(x_train, y_train)
y_pred_ridge = ridge_model.predict(x_test)
```

```python
mse_ridge = mean_squared_error(y_test, y_pred_ridge)
r2_ridge = r2_score(y_test, y_pred_ridge)
print(mse_ridge,r2_ridge)
```
