```python
p75=df["credit_score"].quantile(0.75)
p25=df["credit_score"].quantile(0.25)
print(p25,p75)
```

```python
iqr=p75-p25
print(iqr)
```

```python
ul=p75+1.5*iqr
lp=p25-1.5*iqr
print(ul,lp)
```

```python
df[df["credit_score"]>ul].count()
```

```python
df[df["credit_score"]<lp].count()
```

```python
new_df=df[df["credit_score"]>lp]
sns.boxplot(new_df["credit_score"])
```

```python
sns.histplot(new_df["credit_score"],kde=True)
```

```python
p_75=df["age"].quantile(0.75)
p_25=df["age"].quantile(0.25)
iqr2=p_75-p_25
ul2=p_75+1.5*iqr2
lp2=p_25-1.5*iqr2
print(p_75,p_25,iqr2,ul2,lp2)
```

```python
df[df["age"]>ul2].count()
```

```python
new_df2=df[df["age"]<ul2]
sns.boxplot(new_df2["age"])
```

```python
df_cap = df.copy()
df_cap['age'] = np.where(df_cap['age'] > ul2,ul2,df_cap['age'])
```

```python
sns.boxplot(df_cap["age"])
```

```python
sns.histplot(df_cap["age"],kde=True)
```
